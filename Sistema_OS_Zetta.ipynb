```python
import datetime

class OrdemServico:
    def __init__(self, cliente, equipamento, defeito_relatado):
        # Gera um número de O.S. automático baseado na data e hora
        self.id_os = datetime.datetime.now().strftime("%Y%m%d%H%M")
        self.empresa = "Zetta IA e TI"
        self.cliente = cliente
        self.equipamento = equipamento
        self.defeito_relatado = defeito_relatado
        self.servicos_executados = []
        self.pecas_utilizadas = []
        self.status = "Em Análise"

    def adicionar_servico(self, descricao, valor):
        """Adiciona um serviço de mão de obra ao orçamento"""
        self.servicos_executados.append({"descricao": descricao, "valor": valor})

    def adicionar_peca(self, descricao, valor):
        """Adiciona uma peça trocada ao orçamento"""
        self.pecas_utilizadas.append({"descricao": descricao, "valor": valor})

    def calcular_total(self):
        """Soma o valor das peças e dos serviços"""
        total_serv = sum(item['valor'] for item in self.servicos_executados)
        total_pecas = sum(item['valor'] for item in self.pecas_utilizadas)
        return total_serv + total_pecas

    def gerar_resumo(self):
        """Imprime o "recibo" final da Ordem de Serviço"""
        print(f"=== O.S. [{self.id_os}] | {self.empresa} ===")
        print(f"Cliente: {self.cliente}")
        print(f"Equipamento: {self.equipamento}")
        print(f"Relato Inicial: {self.defeito_relatado}")
        print(f"Status Atual: {self.status}")
        print("------------------------------------------")
        print(f"Total a Pagar: R$ {self.calcular_total():.2f}")
        print("==========================================")
```

```python
# 1. O cliente chega com o equipamento
minha_os = OrdemServico(
    cliente="Carlos Silva", 
    equipamento="TV Xion 50 polegadas", 
    defeito_relatado="Não liga, led standby apagado"
)

# 2. Você faz o diagnóstico e o conserto
minha_os.adicionar_servico("Medição de tensões na placa principal", 80.00)
minha_os.adicionar_servico("Microssolda no circuito de alimentação", 150.00)
minha_os.adicionar_peca("Capacitor SMD", 15.50)

# 3. Atualiza o status e gera o orçamento final
minha_os.status = "Aguardando Retirada"
minha_os.gerar_resumo()
```
